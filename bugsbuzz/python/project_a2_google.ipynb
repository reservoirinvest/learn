```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint  # this is needed to integrate the differential equation(s)

# parameters
M = 180.0     # weight of Alan inKgs
c = 0         # drag coefficient is not a constant
h = 42000     # height from which Alan falls in meters

t = np.linspace(0, 4*60+27, 100) # we will create 100 data points

y0 = [0.0, 0.0] # initial velocity and acceleration
```

```python
def Model(y, t, M, gh, ch):
    """Calculates first and second order differentials
    Args:
        y = [x0, x1] in a delta [t0, t1]
        t = time of the data point in seconds (t1)
        M = mass of in kg
        gh = gravity at height h (computed outside)
        ch = air resistance at height h ( computed outside)
    Returns:
        [first-order differential, second-order differential]"""

    G1 = y[1] # dx/dt
    G2 = gh - (ch/M)*abs(y[1])*y[1] # d2x/dt2
                                            
    return [G1, G2]

def gc_compute(h):
    # Constants
    Re = 6371000.0 # Earth's mean radius in meters
    g0 = 9.8 # gravity at the surface of earth in N/m^2

    # Compute g(h) the gravity at height h
    gh = g0*(Re/(Re+h))**2.0

    # Compute the drag co-efficiant at height h
    ch = 0.5*np.exp(-h/6450)
    return (gh, ch)
```

```python
gc = []
for i in range(len(t)):
    gc[i] = gc_compute(t)
gc
```

```python
ydv = odeint(Model, y0, t, h) # distance and velocity
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('time $t$')
plt.ylabel('position $x$ and velocity $v$')
plt.title("Bungee Jump")
plt.grid()
plt.plot(t, ydv)
plt.legend(('distance', 'velocity'), loc=0)
plt.show()
```
