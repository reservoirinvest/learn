```python
#Load modules 
import matplotlib.pyplot as mpp
import numpy as np

def plot_beta_maps(x0, y0, nDiscard=100, N=180, a1=2.9, betaValues = [0.0, 0.05, 0.15,0.3,0.45,0.5]):
    """Function plots population maps with initial given values
    Args:
        (x0) = float for initial population of x
        (y0) = float for initial population of y
        <nDiscard> = int first data points to discard
        <N> = int for total no of generations

        <a1> = float for known alpha value
        <betaValues> = [list of floats between 0.0 and 0.5]
        
    Returns:
        None. Individual time-series graphs for beta value range in two columns
    """
    
    # Define lambda function for Lmap
    Lmap = lambda alpha, z: alpha*z*(1-z)

    # Initialize variables

    numplot = 1 # Starting plot number
    width, height = 16.0, 20.0 # Width and Height of the plot

    # Set up the figure
    mpp.figure (1,(width,height))

    # Compute no of rows needed for subplot
    nRowSubplot = int(len(betaValues)/2.0) # divide beta-value graphs into 2 columns

    # List for beta values
    b_xy = []

    # Data collection 
    for b in betaValues:
        x =[x0] #intialise list x with x0 as the first value 
        y= [y0] #intialise list y with y0 as the first value 
        
        for n in range(N+1):
            x.append((1.0-b)*Lmap(a1, x[n])+ b*Lmap(a1,y[n])) #generate and apend datapoints for x[n] into list x
            y.append((b*Lmap(a1,x[n]))+(1.0 - b)*Lmap(a1,y[n]))#generate and apend datapoints for x[n] into list x
                
        mpp.subplot(nRowSubplot,2,numplot) #row,column,numplot number 
        
        #plot x[n] and y[n] for each beta value
        mpp.plot(1.0*np.array(x[nDiscard+1:N+1]),"bo-",1.0*np.array(y[nDiscard+1:N+1]),"r^-")
        
        #plot straight line to denote fixed point 
        mpp.axhline(y=((a1-1)/a1), color='green')  
        
        #graph display 
        mpp.ylabel("Population") ; mpp.xlabel("N steps")
        mpp.title("Time series for x[blue], y[red], $beta=$ %.3g" %(b,))
        mpp.grid(True)
    
        numplot += 1 # move to the next beta-value curve
        
        #mpp.subplots_adjust(hspace=0.4, wspace= 0.1)                 
        mpp.tight_layout()
        mpp.legend(("x","y","fixed point"),loc = 0) 
        
        #print table of x[n] and y[n] for each beta values for last 5 n values to observe fixed points. 
        print("\nwhen beta =", b)
        for n in range(N-5, N):
            print ("n =", n, "  x[n] =", round(x[n],5),  "    y[n] =", round(y[n],5))

        b_xy.append([b, x, y]) # appends the x, y values for beta

    mpp.show() # Show the figure

    return b_xy # returns the beta, with x, y values
```

```python
#Intialisation of paramters 
x0, y0 = 0.3, 0.3  #initial population values of x[n] and y[n]
a1 = 2.9  #known alpha value 
betaValues = [0.0, 0.05, 0.15,0.3,0.45,0.5] #list of beta values to test 

nDiscard = 100 # transient datapoints to discard 
N = 180 #total no. of generations

# plot the maps
data = plot_beta_maps(x0, y0)
```
