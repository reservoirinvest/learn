```python
# Load modules:
import numpy as np
import matplotlib.pyplot as mpp

# Define function:
Lmap = lambda alpha, z: alpha*z*(1.0 - z)

def get_lmap(alpha, z):
    """Produces alpha*xn(1-xn)
    Arg: 
        (alpha) is a system parameter
        (z) normalized z-value
    Returns:
        alpha*z*(1.0 - z) as a float"""

    return alpha*z*(1.0 - z)
```

```python
# Setting up simulation parameters:

x0, y0 = 0.5, 0.5 # initial value for x and y
aOne, aTwo = 3.95, 3.20 # alpha_1 and alpha_2

bValues = [0.00, 0.05, 0.3, 0.45, 0.5, 0.55, 0.7, 0.9] # setting up parameter b values

nDiscard = 100 # Number of transient iterations discarded

N = 180 # Number of iterations for data collection after transient

nRowSubplot = int(len(bValues)/2.0) # Each row has two columns for bValues

# Set plot layout:
width, height = 16.0, 20.0
mpp.figure(1, (width, height))

# Data Collection
numplot = 1
for b in bValues:
    x = [x0] # create & initialize list x
    y = [y0] # create & initilize list y

    xMy = x0 - y0 # create a difference list
    xydiff = [xMy] # make that into a list

    for n in range (N+1):
        x.append((1.0-b)*Lmap(aOne, x[n]) + b*Lmap(aTwo, y[n]))
        y.append(b*Lmap(aOne, x[n]) + (1.0-b)*Lmap(aTwo, y[n]))
        xydiff.append(x[n+1] - y[n+1])

    # Data visualization
    mpp.subplot(nRowSubplot, 2, numplot)
    mpp.plot(1.0*np.array(x[nDiscard+1:N+1]), "b-", 1.0*np.array(y[nDiscard+1:N+1]), "r-", 1.0*np.array(xydiff[nDiscard+1: N+1]), "k-")
    mpp.ylabel("Population"); mpp.xlabel("N steps")
    mpp.title("time series for x[blue], y[red] and (x-y)[black], $b=$ %.3g" %(b, ))
    numplot+=1

mpp.subplots_adjust(hspace=0.5, wspace=0.3)
mpp.show()
```

Here we investigate a system of <i><b><span style=color:blue>coupled logistic maps</span></b></i> of the form:

### $$x_{n+1}=(1-b)[\alpha_1 x_n(1-x_n)] + b[\alpha_2 y_n(1-y_n)],$$
### $$y_{n+1}=b[\alpha_1 x_n(1-x_n)] + (1-b)[\alpha_2 y_n(1-y_n)],$$

$n=0, 1, 2, 3, \dots$, where $\alpha_1$, $\alpha_2$, and $b$ are the <i><span style=color:red>system parameters</span></i>. Note that $x_n$ and $y_n$ are <i><span style=color:red>normalized populations.</span></i>

```python
# Load modules
import numpy as np
import matplotlib as mpp

# Initialize variables
x, y = 0.5, 0.5 # initial value for x and y
aOne, aTwo = 3.95, 3.20 # alpha_1 and alpha_2

bValues = [0.00, 0.05, 0.3, 0.45, 0.5, 0.55, 0.7, 0.9] # setting up parameter b values

N = 180 # Number of iterations for data collection after transient

nDiscard = 100 # Discard the first nDiscard iterations

# Build functions for x and y
def get_xydiff(b, aOne, aTwo, x, y):

    """generate x values
    Args: 
        (b), (aOne), (aTwo) as system parameters
        (x) as current x
        (y) as current y
    Returns:
        (nextx, nexty, diff_nextxy) values as a tuple"""

    nextx = (1-b)*(aOne*x*(1-x)) + b*aTwo*y*(1-y)
    nexty = b*(aOne*x*(1-x)) + (1-b)*aTwo*y*(1-y)
    diff_nextxy = nextx - nexty

    return (nextx, nexty, diff_nextxy)

bdata = [] # initialize a list for bvalues containing number of samples

for b in bValues: # Iterate through bValues
    ndata = [] # initialize a list for number of samples
    good_data = []
    for n in range(N): # Iterate through number of samples
        next_xydiff = get_xydiff(b, aOne, aTwo, x, y) # Calculate new x, y and diff
        x, y = next_xydiff[:2] # Make the x, y to the new x, y
        ndata.append(next_xydiff) # Append to the dataset

        if n > nDiscard-1: # discard the first nDiscard data
            good_data.append(next_xydiff)

    bdata.append(good_data)
```
